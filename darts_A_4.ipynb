```python
rows_list = []
for i in range(92):
    i=i+8
    dd = pd.read_csv('carga_2015_2020.csv',usecols=[0,i], header=0)
    df=dd
    df = TimeSeries.from_dataframe(df, time_col='data')
    train, val = df.split_after(pd.Timestamp("2019-01"))
    print('==================')
    print(dd.columns[1])
    print('==================')
    #eval_model(NaiveDrift())
    #eval_model(NaiveSeasonal(12))
    #eval_model(FFT())
    #eval_model(ARIMA())
    #eval_model(AutoARIMA())
    #eval_model(ExponentialSmoothing())
    #eval_model(Prophet())
    eval_model(Theta())
    
    thetas = 2 - np.linspace(-10, 10, 150)
    best_mape = float('inf')
    best_theta = 0
    for theta in thetas:
        model = Theta(theta)
        model.fit(train)
        pred_theta = model.predict(12)
        res = mape(val, pred_theta)
        if res < best_mape:
            best_mape = res
            best_theta = theta
    best_theta_model = Theta(best_theta)
    best_theta_model.fit(train)
    pred_best_theta = best_theta_model.predict(24)
    print(pred_best_theta)
    
    fig, ax = plt.subplots()
    train.plot(label='train', ax=ax)
    val.plot(label='true', ax=ax)
    pred_best_theta.plot(label='prediction', ax=ax)
    ax.legend(loc='best')
    pyplot.show()

    print('Best Theta MAPE is: {:.2f}%, with theta = {}.'.format(mape(val, pred_best_theta), best_theta))
   
```
